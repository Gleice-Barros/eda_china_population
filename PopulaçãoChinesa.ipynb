```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
df = pd.read_csv('/content/china_population.csv') 
```

```python
df
```

```python
print(df['Urban Population'])
```

```python
df.rename(columns={'Urban Population': 'UrbPop'}, inplace = True) #renomeando a coluna
```

```python
df.rename(columns={'Urban Pop %': 'UrbPopPercent'}, inplace = True) #renomeando coluna
```

```python
df
```

```python
df['Rural Population'] = df.apply(
    lambda row: row.Population - row.UrbPop, axis=1) #criando um nova coluna com informação de outras colunas
```

```python
df['RuralPop%'] = df.apply(
    lambda row: 100 - row.UrbPopPercent, axis=1)
```

```python
df
```

```python
index_x = df['Year'] #estabelecendo o eixo x
```

```python
index_y = df['Population'] #estabelecendo o eixo y
index_y_f = df['Fertility Rate']
```

```python
y_pos = np.arange(len(index_x))
```

```python
plt.figure(figsize=(15,7)) #definindo tamanho do gráfico 
plt.plot(- y_pos, index_y) #gerando eixo rotacionado
plt.xticks(- y_pos, index_x)
plt.xlabel('Ano', fontsize = 18) #label eixo x
plt.ylabel('População', fontsize = 18)  #label eixo y
plt.title('Crescimento Populacional', fontsize = 22)
```

```python
index_x1 = df['Year'] #criando o eixo x grafico de barras multiplas
```

```python
index_y_rural = df['RuralPop%'] #criando os eixos y grafico de barras multiplas
index_y_urban = df['UrbPopPercent']
```

```python
plt.figure(figsize=(18,7)) #definindo tamanho do gráfico 
plt.bar(- y_pos, index_y_rural, color = 'lightgreen') #gerando eixo rotacionado y
plt.bar(- y_pos, index_y_urban, color = 'lightblue', bottom = index_y_rural)
plt.xticks(- y_pos, index_x1) 
plt.xlabel('Ano', fontsize = 18) #label eixo x
plt.ylabel('População em %', fontsize = 18)  #label eixo y
plt.legend(('Rural', 'Urbana')) #legenda do gráfico
plt.title('Evolução da População Rural e Urbana', fontsize = 20)
```

```python
index_xl = df ['Year'] #criando o eixo x do grafico de linhas 
```

```python
index_yl = df ['Population'] #criando os eixos y do grafico de linahs
index_yli = df ['Fertility Rate'] 
```

```python
y_pos = np.arange(len(index_xl)) #definindo o arranjo do grafico pelo eixo x
```

```python
plt.figure(figsize=(18,7)) #definindo tamanho do gráfico 

plt.title("Crescimento Populacional x Taxa de Fertilidade", fontsize = 20) 
plt.xticks(- y_pos, index_xl)
plt.plot(- y_pos, index_yl, color='orange', linewidth=3.0) # criando o eixo y com cor e espessura da linha
plt.ylabel("Crescimento populacional (em Bi)", color='orange', fontsize = 12) #colocando rotulo no eixo y e cor
plt.tick_params(axis ='y', labelcolor = 'orange')  #alterando a cor do descritivo do eixo y

ax2 = plt.twinx() #duplicando o eixo y 
plt.plot(- y_pos, index_yli, color='green', linewidth=3.0)
plt.ylabel("Taxa de Fertilidade em %", color = 'green', fontsize = 12)
plt.tick_params(axis ='y', labelcolor = 'green')  
```
